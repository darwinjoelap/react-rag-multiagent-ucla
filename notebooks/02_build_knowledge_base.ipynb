```python
# Cell 1: Imports y Setup
import sys
from pathlib import Path

# Agregar backend al path
backend_path = Path('../backend').resolve()
sys.path.insert(0, str(backend_path))

print(f"âœ… Backend path agregado: {backend_path}")

from app.services.document_loader import DocumentLoader
from app.services.embeddings import EmbeddingService
from pathlib import Path
import logging

# Configurar logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

print("âœ… Imports completados")

# Cell 2: Inicializar servicios
print("Inicializando servicios...")

# Document Loader
doc_loader = DocumentLoader(
    chunk_size=1000,
    chunk_overlap=200
)

# Embedding Service
embedding_service = EmbeddingService()

print(f"âœ… Servicios inicializados")
print(f"ðŸ“Š Modelo de embeddings: {embedding_service.get_model_info()}")

# Cell 3: Verificar directorio de datos
data_dir = Path("../data/raw")
print(f"ðŸ“ Directorio de datos: {data_dir.absolute()}")
print(f"   Existe: {data_dir.exists()}")

if data_dir.exists():
    pdf_files = list(data_dir.glob("*.pdf"))
    print(f"   PDFs encontrados: {len(pdf_files)}")
    for pdf in pdf_files:
        print(f"   - {pdf.name}")
else:
    print("âš ï¸  Directorio no existe. CrÃ©alo y agrega PDFs.")

# Cell 4: Cargar UN PDF de prueba
# Reemplaza 'ejemplo.pdf' con el nombre de tu PDF
test_pdf = "../data/raw/ejemplo.pdf"

print(f"Cargando PDF: {test_pdf}")
documents = doc_loader.load_document(test_pdf)

print(f"\nâœ… PDF cargado exitosamente!")
print(f"   Total de chunks: {len(documents)}")
print(f"   TamaÃ±o promedio: {sum(len(doc.page_content) for doc in documents) / len(documents):.0f} caracteres")

# Cell 5: Inspeccionar chunks
print("ðŸ“„ Primeros 3 chunks:\n")
for i, doc in enumerate(documents[:3], 1):
    print(f"--- Chunk {i} ---")
    print(f"Metadata: {doc.metadata}")
    print(f"Contenido (primeros 200 chars):")
    print(doc.page_content[:200] + "...")
    print()

# Cell 6: Generar embeddings para algunos chunks
print("Generando embeddings de prueba...")

# Tomar primeros 5 chunks
sample_texts = [doc.page_content for doc in documents[:5]]
embeddings = embedding_service.embed_texts(sample_texts)

print(f"âœ… Embeddings generados!")
print(f"   NÃºmero de embeddings: {len(embeddings)}")
print(f"   DimensiÃ³n: {len(embeddings[0])}")
print(f"   Primer embedding (primeros 10 valores):")
print(embeddings[0][:10])

# Cell 7: Probar similitud
print("\nðŸ” Probando similitud entre chunks:")

query = "Â¿QuÃ© es inteligencia artificial?"
query_embedding = embedding_service.embed_query(query)

print(f"\nQuery: '{query}'")
print("\nSimilitud con chunks:")
for i, (doc, emb) in enumerate(zip(documents[:5], embeddings), 1):
    similarity = embedding_service.cosine_similarity(query_embedding, emb)
    print(f"\nChunk {i} (similitud: {similarity:.4f}):")
    print(doc.page_content[:150] + "...")

# Cell 8: Cargar todos los PDFs del directorio
print("Cargando todos los PDFs del directorio...")

all_documents = doc_loader.load_directory("../data/raw")

print(f"\nâœ… Procesamiento completo!")
print(f"   Total de documentos cargados: {len(all_documents)}")
print(f"   Tipos de archivo procesados:")

file_types = {}
for doc in all_documents:
    ftype = doc.metadata.get('file_type', 'unknown')
    file_types[ftype] = file_types.get(ftype, 0) + 1

for ftype, count in file_types.items():
    print(f"   - {ftype}: {count} chunks")

# Cell 9: EstadÃ­sticas
import matplotlib.pyplot as plt

chunk_lengths = [len(doc.page_content) for doc in all_documents]

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.hist(chunk_lengths, bins=30, edgecolor='black')
plt.xlabel('Longitud del chunk (caracteres)')
plt.ylabel('Frecuencia')
plt.title('DistribuciÃ³n de longitud de chunks')

plt.subplot(1, 2, 2)
plt.boxplot(chunk_lengths)
plt.ylabel('Longitud del chunk (caracteres)')
plt.title('Box plot de longitudes')

plt.tight_layout()
plt.show()

print(f"\nðŸ“Š EstadÃ­sticas de chunks:")
print(f"   Media: {sum(chunk_lengths)/len(chunk_lengths):.0f} caracteres")
print(f"   MÃ­nimo: {min(chunk_lengths)} caracteres")
print(f"   MÃ¡ximo: {max(chunk_lengths)} caracteres")
```
