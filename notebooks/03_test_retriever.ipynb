```python
# Cell 1: Setup
import sys
sys.path.insert(0, '../backend')

from app.services.retriever import RetrieverService
from app.services.vector_store import VectorStoreService
import matplotlib.pyplot as plt
import numpy as np

print("âœ… Imports completados")

# Cell 2: Inicializar retriever
retriever = RetrieverService(
    top_k=5,
    similarity_threshold=0.6
)

stats = retriever.get_stats()
print("ðŸ“Š EstadÃ­sticas del sistema:")
for key, value in stats.items():
    print(f"  â€¢ {key}: {value}")

# Cell 3: Consultas de prueba
queries = [
    "Â¿QuÃ© es inteligencia artificial?",
    "machine learning",
    "neural networks",
    "natural language processing",
    "computer vision",
    "deep learning algorithms",
    "transformer models",
    "reinforcement learning"
]

print(f"ðŸ“ Preparadas {len(queries)} consultas de prueba")

# Cell 4: Probar cada consulta
for query in queries[:3]:
    print(f"\n{'='*60}")
    print(f"Query: '{query}'")
    print('='*60)
    
    results = retriever.retrieve(query, top_k=3)
    
    for i, result in enumerate(results, 1):
        print(f"\n[{i}] Similitud: {result['similarity']:.4f}")
        print(f"    Fuente: {result['metadata'].get('source', 'N/A')}")
        print(f"    Tipo: {result['metadata'].get('file_type', 'N/A')}")
        print(f"    Contenido:")
        print(f"    {result['document'][:200]}...")

# Cell 5: AnÃ¡lisis de cobertura
print("\nðŸ“Š ANÃLISIS DE COBERTURA\n")

coverage_results = []
for query in queries:
    coverage = retriever.analyze_query_coverage(query, top_k=10)
    coverage_results.append(coverage)
    
    print(f"Query: '{query}'")
    print(f"  â€¢ Cobertura: {coverage['coverage']}")
    print(f"  â€¢ Similitud max: {coverage['max_similarity']:.4f}")
    print(f"  â€¢ Similitud avg: {coverage['avg_similarity']:.4f}")
    print(f"  â€¢ Fuentes: {coverage['sources_found']}")
    print()

# Cell 6: Visualizar cobertura
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# GrÃ¡fico 1: Similitud mÃ¡xima por query
max_sims = [c['max_similarity'] for c in coverage_results]
axes[0, 0].barh(range(len(queries)), max_sims, color='steelblue')
axes[0, 0].set_yticks(range(len(queries)))
axes[0, 0].set_yticklabels([q[:30] + '...' if len(q) > 30 else q for q in queries], fontsize=8)
axes[0, 0].set_xlabel('Similitud MÃ¡xima')
axes[0, 0].set_title('Mejor Match por Query')
axes[0, 0].axvline(x=0.7, color='red', linestyle='--', label='Threshold')
axes[0, 0].legend()

# GrÃ¡fico 2: Similitud promedio
avg_sims = [c['avg_similarity'] for c in coverage_results]
axes[0, 1].barh(range(len(queries)), avg_sims, color='coral')
axes[0, 1].set_yticks(range(len(queries)))
axes[0, 1].set_yticklabels([q[:30] + '...' if len(q) > 30 else q for q in queries], fontsize=8)
axes[0, 1].set_xlabel('Similitud Promedio')
axes[0, 1].set_title('Similitud Promedio por Query')

# GrÃ¡fico 3: Fuentes encontradas
sources = [c['sources_found'] for c in coverage_results]
axes[1, 0].bar(range(len(queries)), sources, color='lightgreen')
axes[1, 0].set_xticks(range(len(queries)))
axes[1, 0].set_xticklabels(range(1, len(queries)+1))
axes[1, 0].set_ylabel('NÃºmero de Fuentes')
axes[1, 0].set_xlabel('Query #')
axes[1, 0].set_title('Fuentes Ãšnicas por Query')

# GrÃ¡fico 4: DistribuciÃ³n de cobertura
coverage_levels = [c['coverage'] for c in coverage_results]
coverage_counts = {}
for level in coverage_levels:
    coverage_counts[level] = coverage_counts.get(level, 0) + 1

axes[1, 1].pie(
    coverage_counts.values(),
    labels=coverage_counts.keys(),
    autopct='%1.1f%%',
    colors=['#ff6b6b', '#ffd93d', '#6bcf7f', '#4ecdc4']
)
axes[1, 1].set_title('DistribuciÃ³n de Niveles de Cobertura')

plt.tight_layout()
plt.show()

# Cell 7: Comparar recuperaciÃ³n normal vs diversa
query_test = "artificial intelligence and machine learning"
print(f"\nðŸ” Comparando estrategias de recuperaciÃ³n")
print(f"Query: '{query_test}'\n")

# Normal
normal_results = retriever.retrieve(query_test, top_k=5)
print("ðŸ“Œ RecuperaciÃ³n NORMAL:")
for i, r in enumerate(normal_results, 1):
    print(f"  {i}. Similitud: {r['similarity']:.4f} - {r['document'][:80]}...")

# Diversa
diverse_results = retriever.retrieve_diverse(query_test, top_k=5, diversity_factor=0.7)
print("\nðŸŽ¯ RecuperaciÃ³n DIVERSA:")
for i, r in enumerate(diverse_results, 1):
    print(f"  {i}. Similitud: {r['similarity']:.4f} - {r['document'][:80]}...")

# Cell 8: Generar contexto para LLM
query_llm = "Explica los fundamentos de machine learning"
print(f"\nðŸ¤– Generando contexto para LLM")
print(f"Query: '{query_llm}'\n")

context = retriever.get_relevant_context(query_llm, max_tokens=1000)
print(f"Contexto generado: {len(context)} caracteres\n")
print("="*70)
print(context)
print("="*70)

# Cell 9: Probar con diferentes umbrales
query_threshold = "neural networks"
thresholds = [0.5, 0.6, 0.7, 0.8, 0.9]

print(f"\nðŸ“Š Efecto del umbral de similitud")
print(f"Query: '{query_threshold}'\n")

threshold_results = []
for threshold in thresholds:
    retriever_temp = RetrieverService(similarity_threshold=threshold)
    results = retriever_temp.retrieve(query_threshold, top_k=10)
    threshold_results.append(len(results))
    print(f"  Threshold {threshold}: {len(results)} resultados")

# Visualizar
plt.figure(figsize=(10, 5))
plt.plot(thresholds, threshold_results, marker='o', linewidth=2, markersize=8)
plt.xlabel('Umbral de Similitud')
plt.ylabel('NÃºmero de Resultados')
plt.title(f'Resultados vs Umbral - Query: "{query_threshold}"')
plt.grid(True, alpha=0.3)
plt.show()

# Cell 10: Resumen final
print("\n" + "="*70)
print("  RESUMEN FINAL")
print("="*70)

final_stats = retriever.get_stats()
print(f"\nðŸ“Š ConfiguraciÃ³n del Retriever:")
print(f"  â€¢ Top K por defecto: {final_stats['default_top_k']}")
print(f"  â€¢ Umbral de similitud: {final_stats['similarity_threshold']}")
print(f"  â€¢ Total documentos en DB: {final_stats['total_documents']}")
print(f"  â€¢ DimensiÃ³n embeddings: {final_stats['embedding_dimension']}")

print(f"\nâœ… Notebook de testing completado exitosamente!")
```
